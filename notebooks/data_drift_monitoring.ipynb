```python

datastore=Datastore.get(ws, "modeldata")
datastore
```

```python

# specify partition format
partition_format = f'/{ws.subscription_id}/{ws.resource_group}/{ws.name}/aks-service-2/best_model/default/inputs'+'/{date:yyyy/MM/dd}/inputs.csv'
ds = Dataset.Tabular.from_delimited_files(
            datastore.path(path_X_all),
            partition_format=partition_format
        )
        
target = ds.with_timestamp_columns('$aml_dc_scoring_timestamp')

# register the target dataset
target = target.register(ws, 'target-1',create_new_version=True)
# retrieve the dataset from the workspace by name
target = Dataset.get_by_name(ws, 'target-1')
```

```python
dstore=ws.get_default_datastore()
baseline = Dataset.Tabular.from_delimited_files(dstore.path('training-data/train.csv'))
baseline=baseline
```

```python
from azureml.datadrift import DataDriftDetector, AlertConfiguration
```

```python
alert_config = AlertConfiguration(['giosue.cotugno@studio.unibo.it']) # replace with your email to recieve alerts from the scheduled pipeline after enabling
features=["Pclass", "Age", "Sex", "SibSp", "Parch", "Fare", "Embarked"]
monitor = DataDriftDetector.create_from_datasets(ws, 'titanic-monitor', baseline, target, 
                                                      compute_target='testcot',         # compute target for scheduled pipeline and backfills 
                                                      frequency='Day',                     # how often to analyze target data
                                                      feature_list=features,                    # list of features to detect drift on
                                                      drift_threshold=.6,                 # threshold from 0 to 1 for email alerting
                                                      latency=1,                            # SLA in hours for target data to arrive in the dataset
                                                      alert_config=alert_config)            # email addresses to send alert
```

```python
DataDriftDetector.get_by_name(ws,"titanic-monitor").delete()
```
